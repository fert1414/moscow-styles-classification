```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def add_results(model_name, results, model_result):
    result = {
        'val_loss_history': model_result['val_loss_history'].to_list(),
        'val_accuracy_history': model_result['val_accuracy_history'].to_list(),
        'train_loss_history': model_result['train_loss_history'].to_list(),
        'train_accuracy_history': model_result['train_accuracy_history'].to_list()
    }
    results[model_name] = result

def show_results(results, num_epochs):
    plt.figure(figsize = (25, 20))

    index = 1
    for model in results:
        model_result = results[model]

        plt.subplot(len(results), 2, index)
        plt.plot(range(num_epochs), model_result['val_loss_history'], label = 'val')
        plt.plot(range(num_epochs), model_result['train_loss_history'], label = 'train')

        plt.title(model)
        plt.grid()
        plt.legend()

        plt.subplot(len(results), 2, index + 1)
        plt.plot(range(num_epochs), model_result['val_accuracy_history'], label = 'val')
        plt.plot(range(num_epochs), model_result['train_accuracy_history'], label = 'train')

        plt.title(model)
        plt.grid()
        plt.legend()

        index += 2

    plt.show()
```

### Результаты для resnet18, resnet50, resnet152 на Adam и уменьшением шага градиента на 0.1 через каждые 7 эпох

```python
results_resnet18 = pd.read_csv('./models_results/resnet18.csv')
results_resnet50 = pd.read_csv('./models_results/resnet50.csv')
results_resnet152 = pd.read_csv('./models_results/resnet152.csv')

results = {}
add_results('resnet18', results, results_resnet18)
add_results('resnet50', results, results_resnet50)
add_results('resnet152', results, results_resnet152)
```

```python
show_results(results, 100)
```

```python
results_resnet152_1 = pd.read_csv('./models_results/resnet152_1.csv')
results_resnet152_2 = pd.read_csv('./models_results/resnet152_2.csv')
results_resnet152_3 = pd.read_csv('./models_results/resnet152_3.csv')

results = {}
add_results('resnet152_1', results, results_resnet152_1)
add_results('resnet152_2', results, results_resnet152_2)
add_results('resnet152_3', results, results_resnet152_3)
```

```python
show_results(results, 20)
```
